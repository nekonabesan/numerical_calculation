### ⾵⼒発電プロジェクトの正味現在価値（NPV）の計算

$ O\&M costs (euro / year ) = AEP \times O\&M costs $

$ AR = AEP \times ESP $

$ ANI = AR - O\&M costs (euro / year ) $

$ NPV = \Sigma_{P = 1}^{20} - CI, $ for using function $ NPV: NPV(R(ANI;ANI:...xP)) - CI $

$ PVM = NPV(R(O\&M costs ( euro / year ))); O\&M costs ( euro / year );..xP $

$ PV = FV \displaystyle\frac{1}{(1 + R)^{P}}, $ for function $ NPV: PVDC = PV(R;P;;DC) $

$ PVC = PVDC + PVM + CI $

$ LAC:PMT = \displaystyle\left[ \displaystyle\frac{-PVC \times R}{ 1 - (1 + R)^{-P} } \right], $  for using function $ PMT: PMT(R;P;PVC) $

$ LCoE = \displaystyle\frac{LAC}{AEP} $

但し

- AR - 電力販売による年間収益（€/年）
- AEP - 年間発電量（kWh）
- ESP - 電力販売価格（€/kWh）
- ANI - 年間純利益（€/年）
- NPV - 正味現在価値（€）
- CI - 設備投資額（€）
- R - 年間割引率（%）
- P - 年数（年）
- DC - 廃止措置費用（€）
- PVDC - 廃止措置費用の現在価値（€）
- PVC - 全費用の現在価値（€）
- LAC - 年間均等化発電原価（€）
- LCoE - 均等化発電原価（€/kWh）
- O&M費用 - 年間運転・保守費用（€/kWh）
- PVM - 保守・修理費用の現在価値
- FV - 将来価値
- PV - 将来における金額の現在価値


NPV関数の構⽂は以下のとおりです
```
NPV(rate,value1,value2,...(last value- decommissioning cost)) - initial capital investment
```

但し

- `rate` : 1年間の割引率
- `value1` : 1年⽬の年間純利益
- `value2` : 2年⽬の年間純利益（以降最終年の1年前まで同様）
- `last value` 、: 前年度における年間純利益
- `decommissioning cost`: 放射性物質の除去と段階的な工場の解体を含む、行政的かつ技術的なプロセス
- `initial capital investment`: 0年目の初めに行われる初期資本投資

NPVを計算するには、ほとんどのスプレッドにある「NPV」関数を使⽤できます。

### ⾵⼒発電所からの均等化発電原価（LCOE）の計算

プロジェクトの20年間のライフサイクルにおける均等化発電原価を計算する前に、
すべてのコストの現在価値を計算する必要があります（すなわち、初期資本コスト、
プロジェクトライフサイクルにおける運用・保守（O&M）コスト、および
廃止措置コスト）。

O&Mコストの正味現在価値はNPV関数で計算されます。

廃止措置コストの現在価値は、PV関数を使用して計算します。

PV関数の構文は以下のとおりです。

```
PV(rate, number of periods, annual payment, future value),
```

但し
- `rate`: 1年間の割引率
- `number of periods`: プロジェクトの存続期間（年）
- `annual payment` is zero
- `future value`: 風力発電所の廃止措置にかかる費用はいくらですか？

正味現在価値（NPV）を計算するには、ほとんどの表計算ソフトに搭載されている「NPV」関数を使用できます。

均等化年間コストを計算するには、PMT関数（別名：支払い関数）を使用できます。構文は以下のとおりです。

但し
- `rate`: 1年間の割引率
- `number of periods`: プロジェクトの存続期間（年）
- `present value`: すべての費用の合計の現在価値

質問４

この風力発電プロジェクトの正味現在価値をもとめなさい(単位はユーロ/年、絶対値で回答すること)

````python
# 質問4: 風力発電プロジェクトの正味現在価値 (NPV)
aep = 110_000_000  # kWh/year
esp = 0.075  # euro/kWh
om_cost_per_kwh = 0.0091  # euro/kWh
ci = 65_000_000  # euro
dc = 3_000_000  # euro
r = 0.08
p = 20

annual_revenue = aep * esp
om_cost_per_year = aep * om_cost_per_kwh
annual_net_income = annual_revenue - om_cost_per_year

# 最終年は廃止費用を差し引く
cashflows = [annual_net_income] * (p - 1) + [annual_net_income - dc]

npv = sum(cf / ((1 + r) ** year) for year, cf in enumerate(cashflows, start=1)) - ci

print(f"NPV: {abs(npv):,.0f} €")
print(f"(符号付きNPV: {npv:,.0f} €)")
````

## 風力発電の経済性 - 平準化エネルギーコスト(LCoE)

割引率を0.08とする

質問１

運転開始後20年間の保守費用の現在価値を求めなさい(単位はユーロ、絶対値で回答すること)

````python
# LCoE 質問1: O&M費用の現在価値 (PVM)
r = 0.08
p = 20
om_cost_per_year = 1_001_000
pvm = sum((-om_cost_per_year) / ((1 + r) ** year) for year in range(1, p + 1))
print(f"20年間の保守費用の現在価値: {abs(pvm):,.0f} €")
````

質問２

風力発電所を廃止するために支払う費用の現在価値を求めなさい(単位はユーロ、絶対値で回答すること)

````python
# LCoE 質問2: 廃止費用の現在価値 (PVDC)
r = 0.08
p = 20
dc = 3_000_000
pvdc = (-dc) / ((1 + r) ** p)
print(f"廃止費用の現在価値: {abs(pvdc):,.0f} €")
````

質問３

全ての費用の現在価値を求めなさい(単位はユーロ、絶対値で回答すること)

````python
# LCoE 質問3: 全費用の現在価値 (PVC)
r = 0.08
p = 20
ci = 65_000_000
om_cost_per_year = 1_001_000
dc = 3_000_000
pvm = sum((-om_cost_per_year) / ((1 + r) ** year) for year in range(1, p + 1))
pvdc = (-dc) / ((1 + r) ** p)
pvc = ci + abs(pvm) + abs(pvdc)
print(f"全費用の現在価値: {pvc:,.0f} €")
````

質問４

平準化された年間コストを求めなさい(単位はユーロ/年、絶対値で回答すること)

````python
# LCoE 質問4: 平準化年間コスト (LAC)
r = 0.08
p = 20
ci = 65_000_000
om_cost_per_year = 1_001_000
dc = 3_000_000
pvm = sum((-om_cost_per_year) / ((1 + r) ** year) for year in range(1, p + 1))
pvdc = (-dc) / ((1 + r) ** p)
pvc = ci + abs(pvm) + abs(pvdc)
lac = pvc * r / (1 - (1 + r) ** (-p))
print(f"平準化年間コスト: {lac:,.0f} €/年")
````

質問５

風力発電所の平準化エネルギーコスト(LCoE)を求めなさい(単位はユーロ/kWh、絶対値で回答すること)

````python
# LCoE 質問5: 平準化エネルギーコスト (€/kWh)
r = 0.08
p = 20
aep = 110_000_000
ci = 65_000_000
om_cost_per_year = 1_001_000
dc = 3_000_000
pvm = sum((-om_cost_per_year) / ((1 + r) ** year) for year in range(1, p + 1))
pvdc = (-dc) / ((1 + r) ** p)
pvc = ci + abs(pvm) + abs(pvdc)
lac = pvc * r / (1 - (1 + r) ** (-p))
lcoe_eur_per_kwh = lac / aep
print(f"LCoE: {lcoe_eur_per_kwh:.5f} €/kWh")
````

質問６

平準化エネルギーコスト(LCoE)をユーロセント/kWhで表し、絶対値で回答しなさい

````python
# LCoE 質問6: 平準化エネルギーコスト (ユーロセント/kWh)
r = 0.08
p = 20
aep = 110_000_000
ci = 65_000_000
om_cost_per_year = 1_001_000
dc = 3_000_000
pvm = sum((-om_cost_per_year) / ((1 + r) ** year) for year in range(1, p + 1))
pvdc = (-dc) / ((1 + r) ** p)
pvc = ci + abs(pvm) + abs(pvdc)
lac = pvc * r / (1 - (1 + r) ** (-p))
lcoe_eur_per_kwh = lac / aep
lcoe_cent_per_kwh = lcoe_eur_per_kwh * 100
print(f"LCoE: {lcoe_cent_per_kwh:.3f} €c/kWh")
````
